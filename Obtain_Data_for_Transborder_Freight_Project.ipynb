## Identify csv files after extraction

Identify all CSV files within the extracted_data/csv_files directory and its subdirectories.



```python
import os

extracted_csv_dir = 'extracted_data/csv_files'
csv_files = []

# Walk through the directory and its subdirectories
for root, dirs, files in os.walk(extracted_csv_dir):
    for file in files:
        if file.endswith(".csv"):
            csv_files.append(os.path.join(root, file))

# Print the list of identified CSV files
print("Identified CSV files:")
for csv_file in csv_files:
    print(csv_file)
```

## Load and combine data

Load each CSV file into a pandas DataFrame and combine them into a single DataFrame.


```python
dfs = []

for csv_file in csv_files:
    # Skip files in the __MACOSX directory
    if '__MACOSX' in csv_file:
        continue
    try:
        # Set low_memory to False to avoid DtypeWarning
        df = pd.read_csv(csv_file, low_memory=False)
        dfs.append(df)
    except Exception as e:
        print(f"Error reading {csv_file}: {e}")

combined_df = pd.concat(dfs, ignore_index=True)
```

Display the head of the combined DataFrame to verify the concatenation was successful.



```python
display(combined_df.head())
```

## Save combined DataFrame to CSV

Save the combined DataFrame to a CSV file for future use, so that I don't have to repeat the extraction and merging process.

```python
# Save the combined DataFrame to a CSV file
combined_df.to_csv('combined_data.csv', index=False)

print("Combined DataFrame saved to combined_data.csv")
```

## Delete the original zip split files and files

Delete the original zip split files and files, so that I can free up some space.

```python
import os
import shutil

# List of files to remove
files_to_remove = [
    'ezyZip.zip',
    'merged.zip'
]

# Remove the original split zip files
for i in range(1, 8):
    files_to_remove.append(f'ezyZip.z{i:02d}')

# Directories to remove
dirs_to_remove = [
    'extracted_files', # This contains the original data.zip
    'extracted_data/data' # This contains the nested zip files
]

# Remove files
for file_path in files_to_remove:
    if os.path.exists(file_path):
        os.remove(file_path)
        print(f"Removed file: {file_path}")

# Remove directories
for dir_path in dirs_to_remove:
    if os.path.exists(dir_path):
        shutil.rmtree(dir_path)
        print(f"Removed directory: {dir_path}")
```
